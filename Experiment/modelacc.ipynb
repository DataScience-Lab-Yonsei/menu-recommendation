```python
# Top 3
a = []
 
for i in range(4532):
    b = pd.DataFrame(pred_df[i].sort_values(ascending=False))
    line = list(b.iloc[:3,].index)
    a.append(line)         

print(a)
```

```python
# Answer to df
b = np.nonzero(true)
b = pd.DataFrame(b)
b = b.drop([0])
b = b.transpose()
b.columns = ["answer"]
```

```python
# Concat[Top3, answer]
gru_output = pd.concat([a,b], axis = 1)
```

## Category & Property col 추가

```python
data = pd.read_csv("/content/drive/MyDrive/data/total_menu_final_category_property.csv")
```

```python
key = []
value = []
for i in range(2529):
  cat = np.unique(data[data["menu_id"] == i]["category"])
  key.append(i)
  value.append(cat)
```

```python
# Category df
d = dict({'menu_id': key, 'category' : value})
d = pd.DataFrame(d)
```

```python
key1 = []
value1 = []
for i in range(2529):
  ppt = np.unique(data[data["menu_id"] == i]["property"])
  key1.append(i)
  value1.append(ppt)
```

```python
# Property df
p = dict({'menu_id': key1, 'property' : value1})
p = pd.DataFrame(p)
```

## 정답 추가 

```python
pd.set_option('mode.chained_assignment',  None) 
```

```python
# answer dataframe
ans = pd.concat([d,p['property']], axis = 1)
```

```python
gru_output['category'] = 0
gru_output['property'] = 0

answer_id = list(gru_output['answer'])
menu_id = list(ans['menu_id'])
ctg = list(ans['category'])
pp = list(ans['property'])
```

```python
for i in range(len(menu_id)):
    for j in range(len(answer_id)):
        if gru_output['answer'][j] == menu_id[i]:
            gru_output['category'][j] = ctg[i]
            gru_output['property'][j] = pp[i]
```

```python
# 첫번째 열
first = gru_output[['firstrec']]
first['ctg1'] = 0
first['pp1'] = 0

answer_id = list(first['firstrec'])
```

```python
for i in range(len(menu_id)):
    for j in range(len(answer_id)):
        if first['firstrec'][j] == menu_id[i]:
            first['ctg1'][j] = ctg[i]
            first['pp1'][j] = pp[i]
```

```python
# 두번째 열
second = gru_output[['secondrec']]
second['ctg2'] = 0
second['pp2'] = 0

answer_id = list(second['secondrec'])

for i in range(len(menu_id)):
    for j in range(len(answer_id)):
        if second['secondrec'][j] == menu_id[i]:
          second['ctg2'][j] = ctg[i]
          second['pp2'][j] = pp[i]
```

```python
# 세번째 열
third = gru_output[['thirdrec']]
third['ctg3'] = 0
third['pp3'] = 0

answer_id = list(third['thirdrec'])

for i in range(len(menu_id)):
    for j in range(len(answer_id)):
        if third['thirdrec'][j] == menu_id[i]:
          third['ctg3'][j] = ctg[i]
          third['pp3'][j] = pp[i]
```

# Category hit 지표

```python
category_hit = []
for i in range(len(first)):
  if gru_output['category'][i] == first['ctg1'][i]:
    category_hit.append(1)
  elif gru_output['category'][i] == second['ctg2'][i]:
    category_hit.append(0.8)
  elif gru_output['category'][i] == third['ctg3'][i]:
    category_hit.append(0.6)
  else: category_hit.append(0)
```

```python
gru_output['category_hit'] = category_hit
```

# Property hit 지표

```python
property_hit = []
for i in range(len(first)):
  if (gru_output['property'][i] == first['pp1'][i]).all():
    property_hit.append(1)
  elif (gru_output['property'][i] == second['pp2'][i]).all():
    property_hit.append(0.8)
  elif (gru_output['property'][i] == third['pp3'][i]).all():
    property_hit.append(0.6)
  else: property_hit.append(0)
```

```python
gru_output['property_hit'] = property_hit
```

# Hit 지표
- 세 개 중에 하나라도 정답이 있으면 1 부여

```python
hit = []
for i in range(len(first)):
  if gru_output['answer'][i] == first['firstrec'][i]:
    hit.append(1)
  elif gru_output['answer'][i] == second['secondrec'][i]:
    hit.append(1)
  elif gru_output['answer'][i] == third['thirdrec'][i]:
    hit.append(1)
  else: hit.append(0)
```

```python
gru_output['hit'] = hit
```

```python
# HIT@3 지표 계산
HIT = sum(gru_output['hit'])/len(gru_output)
HIT
```
