```python
%cd /content/drive/MyDrive/data/result
```

```python
df_rec = pd.read_csv('rec_ctg_pp.csv')
df_rec
```

```python
user = df_rec['user'].unique()
menu_list = list(df_rec['item'])
ctg = list(df_rec['category'])
pp = list(df_rec['property'])
```

```python
idx0 = list(range(0, len(menu_list), 3))
idx1 = list(range(1, len(menu_list), 3))
idx2 = list(range(2, len(menu_list), 3))
```

```python
# firstrec 생성
firstrec = []
ctg1 = []
pp1 = []
for idx in idx0:
  firstrec.append(menu_list[idx])
  ctg1.append(ctg[idx])
  pp1.append(pp[idx])
```

```python
first = pd.DataFrame([user,firstrec,ctg1,pp1], index = ['user','firstrec','ctg1','pp1'])
first = first.transpose()
```

```python
# secondrec 생성
secondrec = []
ctg2 = []
pp2 = []
for idx in idx1:
  secondrec.append(menu_list[idx])
  ctg2.append(ctg[idx])
  pp2.append(pp[idx])
```

```python
second = pd.DataFrame([user,secondrec,ctg2,pp2], index = ['user','secondrec','ctg2','pp2'])
second = second.transpose()
```

```python
# thirdrec 생성
thirdrec = []
ctg3 = []
pp3 = []
for idx in idx2:
  thirdrec.append(menu_list[idx])
  ctg3.append(ctg[idx])
  pp3.append(pp[idx])
```

```python
third = pd.DataFrame([user,thirdrec,ctg3,pp3], index = ['user','thirdrec','ctg3','pp3'])
third = third.transpose()
```

# 원본 데이터 전처리

## 정답값만 뽑기

```python
origin = pd.read_csv('/content/drive/Shareddrives/DSL_Modeling_B/result/origin.csv')
```

```python
user = origin['user'].unique()
```

```python
org = origin.set_index('user')
```

```python
answer_org = []

for i in range(len(user)):
  lst = org.loc[user[i]]['item'].tolist()
  answer_org.append(lst[-1])
```

```python
true = pd.DataFrame([user,answer_org ], index = ['user','Menu']).transpose()
```

```python
true
```

## category, property 부여

```python
true['category'] = 0
true['property'] = 0
```

```python
cp = pd.read_csv('/content/drive/Shareddrives/DSL_Modeling_B/data/dacon_menu_final_category_property.csv')
```

```python
cp
```

```python
o_menu = cp['menu'].tolist()
o_menu_id = cp['menu_id'].tolist()
o_category = cp['category'].tolist()
o_property = cp['property'].tolist()
```

```python
true['menu_id'] = 0
```

```python
for i in range(len(o_menu)):
    for j in range(len(true)):
        if true['Menu'][j] == o_menu[i]:
            true['category'][j] = o_category[i]
            true['property'][j] = o_property[i]
            true['menu_id'][j] = o_menu_id[i]    
```

```python
true
```

```python
true.to_csv("/content/drive/Shareddrives/DSL_Modeling_B/result_true.csv")
```

# Category hit 지표

```python
category_hit = []
for i in range(len(true)):
  if true['category'][i] == first['ctg1'][i]:
    category_hit.append(1)
  elif true['category'][i] == second['ctg2'][i]:
    category_hit.append(0.8)
  elif true['category'][i] == third['ctg3'][i]:
    category_hit.append(0.6)
  else: category_hit.append(0)
```

# Property hit 지표

```python
property_hit = []
for i in range(len(true)):
  if true['property'][i] == first['pp1'][i]:
    property_hit.append(1)
  elif true['property'][i] == second['pp2'][i]:
    property_hit.append(0.8)
  elif true['property'][i] == third['pp3'][i]:
    property_hit.append(0.6)
  else: property_hit.append(0)
```

# Hit 지표

```python
hit = []
for i in range(len(true)):
  if true['menu_id'][i] == first['firstrec'][i]:
    hit.append(1)
  elif true['menu_id'][i] == second['secondrec'][i]:
    hit.append(1)
  elif true['menu_id'][i] == third['thirdrec'][i]:
    hit.append(1)
  else: hit.append(0)
```

# 결과 확인

```python
output = pd.DataFrame([user,first['firstrec'], second['secondrec'], third['thirdrec'], true['Menu'], category_hit, property_hit, hit], 
                      index = ['user','first','second','third','answer',"category_hit", "property_hit", "hit"]).transpose()
```

```python
output = output.dropna()
```
